```python
X = df.drop("Outcome",axis=1)
y = df["Outcome"]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
model = svm.SVC(kernel="linear")
```

```python
model.fit(X_train,y_train)
```

```python
score_train = model.predict(X_train)
accuracy_score(y_train,score_train)
```

```python
score_test = model.predict(X_test)
accuracy_score(y_test,score_test)
```

```python
def predicting():
    input_data = (4,110,92,0,0,37.6,0.191,30)
    input_aray = np.asarray(input_data)
    input_aray = input_aray.reshape(1,-1)
    input_aray = sc.transform(input_aray)
    print(input_aray)
    predic = model.predict(input_aray)
    
    if(predic[0]==0):
        print("The person is Non diabetic")
    else:
        print("The person is Diabetic")
```
